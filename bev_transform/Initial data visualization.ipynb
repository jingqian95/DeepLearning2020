```python
import os
import random
import math

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = [5, 5]
matplotlib.rcParams['figure.dpi'] = 200
```

```python
anno = pd.read_csv('data/annotation_newfeat.csv')
```

```python
anno.columns
```

```python
wh_ratio = anno.apply(lambda x: max(x['box_width'],x['box_height'])/min(x['box_width'],x['box_height']),1)
plt.hist(wh_ratio)
plt.title('Width/Height ratio of bounding boxes')
plt.xlabel('w/h ratio')
plt.ylabel('frequency')
```

```python
area = anno['box_width']*anno['box_height']
plt.hist(area)
plt.title('Area of bounding boxes')
plt.xlabel('area')
plt.ylabel('frequency')
```
