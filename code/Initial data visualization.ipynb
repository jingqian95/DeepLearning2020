```python
import os
import random
import math

import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = [5, 5]
matplotlib.rcParams['figure.dpi'] = 200
```

```python
anno = pd.read_csv('../datasets/dl2020/annotation_newfeat_2.csv')
```

```python
anno.columns
```

```python
wh_ratio = anno.apply(lambda x: max(x['box_width'],x['box_height'])/min(x['box_width'],x['box_height']),1)
plt.figure(figsize = (3,3))
plt.hist(wh_ratio, bins = 30)
plt.title('Width/Height ratio of bounding boxes')
plt.xlabel('w/h ratio')
plt.ylabel('frequency')
```

```python
area = anno['box_width']*anno['box_height']
plt.hist(area)
plt.title('Area of bounding boxes')
plt.xlabel('area')
plt.ylabel('frequency')
```
